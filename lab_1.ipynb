```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("C:\\Users\\abhi\\Documents\\GitHub\\Data-Visualization-Lab-CSVTU\\Dataset\\train.csv")

# Display dataset information
print(df.info())
print(df.head())
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["OverallQual"], y=df["SalePrice"], palette="coolwarm")
plt.title("House Price Distribution by Overall Quality")
plt.xlabel("Overall Quality")
plt.ylabel("Sale Price")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["ExterQual"], y=df["SalePrice"], order=['Po', 'Fa', 'TA', 'Gd', 'Ex'], palette="viridis")
plt.title("House Prices by Exterior Quality")
plt.xlabel("Exterior Quality")
plt.ylabel("Sale Price")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["GrLivArea"], y=df["SalePrice"], hue=df["OverallQual"], size=df["OverallCond"], palette="coolwarm")
plt.title("Living Area vs Sale Price")
plt.xlabel("Above Ground Living Area (sqft)")
plt.ylabel("Sale Price")
plt.show()
```

```python
   plt.figure(figsize=(10, 6))
sns.histplot(df["SalePrice"], bins=50, kde=True, color="blue")
plt.title("Distribution of House Prices")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["GrLivArea"], y=df["SalePrice"], hue=df["OverallQual"], size=df["TotalBsmtSF"], palette="coolwarm")
plt.title("Living Area vs Sale Price (Colored by Quality)")
plt.xlabel("Above Ground Living Area (sqft)")
plt.ylabel("Sale Price")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["KitchenQual"], y=df["SalePrice"], order=['Po', 'Fa', 'TA', 'Gd', 'Ex'], palette="mako")
plt.title("Sale Price by Kitchen Quality")
plt.xlabel("Kitchen Quality")
plt.ylabel("Sale Price")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df["YearBuilt"], y=df["SalePrice"], estimator="median", ci=None, color="red")
plt.title("House Prices Over the Years")
plt.xlabel("Year Built")
plt.ylabel("Median Sale Price")
plt.show()
```

```python
sns.pairplot(df[["SalePrice", "GrLivArea", "TotalBsmtSF", "OverallQual"]], hue="OverallQual", palette="coolwarm")
plt.show()
```
